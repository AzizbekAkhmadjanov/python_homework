1. Write a Python program to handle a ZeroDivisionError exception when dividing a number by zero.



```python
try:
    numerator = float(input("Enter numerator: "))
    denominator = float(input("Enter denominator: "))
    
    result = numerator / denominator
    print("Result:", result)

except ZeroDivisionError:
    print("Error: Cannot divide by zero!")

except ValueError:
    print("Error: Please enter a valid number.")
```

```python
def get_number(prompt):
    value = input(prompt)
    try:
        return float(value)  # Convert input to a number (int or float)
    except ValueError:
        raise TypeError(f"Invalid input '{value}': not a number.")

try:
    num1 = get_number("Enter the first number: ")
    num2 = get_number("Enter the second number: ")

    print(f"You entered numbers: {num1} and {num2}")

except TypeError as e:
    print("TypeError:", e)
```

1. Write a Python program to read an entire text file.



```python
file_name = input("Enter the file name to read: ")

try:
    with open(file_name, 'r', encoding='utf-8') as file:
        content = file.read()
        print("File content:\n")
        print(content)

except FileNotFoundError:
    print(f"Error: The file '{file_name}' does not exist.")

except UnicodeDecodeError:
    print(f"Error: Cannot decode file '{file_name}'. Check its encoding.")
```

2. Write a Python program to read first n lines of a file.



```python
file_name = input("Enter the file name to read: ")
n = int(input("Enter the number of lines to read: "))

try:
    with open(file_name, 'r', encoding='utf-8') as file:
        for i in range(n):
            line = file.readline()
            if not line:  # end of file reached
                break
            print(line, end='')  # `end=''` avoids adding extra newline

except FileNotFoundError:
    print(f"Error: The file '{file_name}' does not exist.")

except UnicodeDecodeError:
    print(f"Error: Cannot decode file '{file_name}'. Check its encoding.")

except ValueError:
    print("Error: Please enter a valid number for n.")
```

3. Write a Python program to append text to a file and display the text.



```python
file_name = input("Enter the file name: ")
text_to_append = input("Enter the text to append: ")

try:
    # Append text to the file
    with open(file_name, 'a', encoding='utf-8') as file:
        file.write(text_to_append + '\n')

    print(f"\nText appended successfully to '{file_name}'.\n")

    # Display the updated file content
    with open(file_name, 'r', encoding='utf-8') as file:
        print("Updated File Content:\n")
        print(file.read())

except FileNotFoundError:
    print(f"Error: The file '{file_name}' does not exist.")

except UnicodeDecodeError:
    print(f"Error: Cannot decode file '{file_name}'. Check its encoding.")
```

4. Write a Python program to read last n lines of a file.



```python
file_name = input("Enter the file name: ")
n = int(input("Enter the number of lines to read from the end: "))

try:
    with open(file_name, 'r', encoding='utf-8') as file:
        lines = file.readlines()  # read all lines into a list
        last_lines = lines[-n:]   # get the last n lines

        print(f"\nLast {n} lines of '{file_name}':\n")
        for line in last_lines:
            print(line, end='')

except FileNotFoundError:
    print(f"Error: The file '{file_name}' does not exist.")

except UnicodeDecodeError:
    print(f"Error: Cannot decode file '{file_name}'. Check its encoding.")

except ValueError:
    print("Error: Please enter a valid number for n.")
```

8. Write a Python program to find the longest words.



```python
text = input("Enter a sentence or paragraph: ")

# Split text into words
words = text.split()

# Find the length of the longest word(s)
max_length = max(len(word) for word in words)

# Find all words with the maximum length
longest_words = [word for word in words if len(word) == max_length]

print(f"\nLongest word length: {max_length}")
print("Longest word(s):", longest_words)
```

10. Write a Python program to count the frequency of words in a file.



```python
from collections import Counter
import string

file_name = input("Enter the file name: ")

try:
    with open(file_name, 'r', encoding='utf-8') as file:
        text = file.read().lower()  # convert to lowercase for uniformity

    # Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))

    # Split text into words
    words = text.split()

    # Count frequency using Counter
    word_counts = Counter(words)

    print(f"\nWord frequencies in '{file_name}':")
    for word, count in word_counts.items():
        print(f"{word}: {count}")

except FileNotFoundError:
    print(f"Error: The file '{file_name}' does not exist.")

except UnicodeDecodeError:
    print(f"Error: Cannot decode file '{file_name}'. Check its encoding.")
```

14. Write a Python program to combine each line from the first file with the corresponding line in the second file.



```python
file1_name = input("Enter the first file name: ")
file2_name = input("Enter the second file name: ")
output_file_name = input("Enter the output file name: ")

try:
    with open(file1_name, 'r', encoding='utf-8') as file1, \
         open(file2_name, 'r', encoding='utf-8') as file2, \
         open(output_file_name, 'w', encoding='utf-8') as outfile:

        # Read lines from both files
        for line1, line2 in zip(file1, file2):
            combined_line = line1.rstrip() + " " + line2  # remove newline from first file
            outfile.write(combined_line)

    print(f"Lines from '{file1_name}' and '{file2_name}' have been combined into '{output_file_name}'.")

except FileNotFoundError as e:
    print(f"Error: {e.filename} does not exist.")

except IOError:
    print("Error: Could not read/write files.")
```

16. Write a Python program to assess if a file is closed or not.



```python
file_name = input("Enter the file name: ")

try:
    file = open(file_name, 'r', encoding='utf-8')
    print(f"File '{file_name}' opened.")
    
    # Check if the file is closed
    print("Is the file closed?", file.closed)

    file.close()  # close the file
    print("File closed.")

    # Check again
    print("Is the file closed now?", file.closed)

except FileNotFoundError:
    print(f"Error: The file '{file_name}' does not exist.")

except IOError:
    print(f"Error: Could not open the file '{file_name}'.")
```

17. Write a Python program to remove newline characters from a file.



```python
file_name = input("Enter the file name: ")

try:
    # Read the file content
    with open(file_name, 'r', encoding='utf-8') as file:
        lines = file.readlines()

    # Remove newline characters and join into a single string
    new_content = ''.join(line.rstrip('\n') for line in lines)

    # Write the modified content back to the file
    with open(file_name, 'w', encoding='utf-8') as file:
        file.write(new_content)

    print(f"All newline characters have been removed from '{file_name}'.")

except FileNotFoundError:
    print(f"Error: The file '{file_name}' does not exist.")

except IOError:
    print(f"Error: Could not read/write the file '{file_name}'.")
```

18. Write a Python program that takes a text file as input and returns the number of words in a given text file.

Note: Some words can be separated by a comma with no space.

```python
import string

file_name = input("Enter the file name: ")

try:
    with open(file_name, 'r', encoding='utf-8') as file:
        text = file.read()

    # Replace punctuation (like , . ! ?) with spaces
    translator = str.maketrans(string.punctuation, ' '*len(string.punctuation))
    clean_text = text.translate(translator)

    # Split on whitespace
    words = clean_text.split()
    word_count = len(words)

    print(f"The file '{file_name}' has {word_count} word(s).")

except FileNotFoundError:
    print(f"Error: The file '{file_name}' does not exist.")

except UnicodeDecodeError:
    print(f"Error: Cannot decode file '{file_name}'. Check its encoding.")
```

19. Write a Python program to extract characters from various text files and put them into a list.



```python
file_name = input("Enter the file name: ")

characters_list = []

try:
    with open(file_name, 'r', encoding='utf-8') as file:
        for line in file:
            # Add each character to the list (including spaces and punctuation)
            characters_list.extend(list(line.rstrip('\n')))  # remove newline

    print(f"Characters from '{file_name}' have been stored in a list:")
    print(characters_list)

except FileNotFoundError:
    print(f"Error: The file '{file_name}' does not exist.")

except UnicodeDecodeError:
    print(f"Error: Cannot decode file '{file_name}'. Check its encoding.")
```

20. Write a Python program to generate 26 text files named A.txt, B.txt, and so on up to Z.txt.



```python
import string

# Loop through uppercase letters A-Z
for letter in string.ascii_uppercase:
    file_name = f"{letter}.txt"
    try:
        with open(file_name, 'w', encoding='utf-8') as file:
            file.write(f"This is file {file_name}\n")  # optional content

        print(f"File '{file_name}' created successfully.")

    except IOError:
        print(f"Error: Could not create file '{file_name}'.")
```

21. Write a Python program to create a file where all letters of the English alphabet are listed with a specified number of letters on each line.



```python
import string

file_name = input("Enter the file name to create: ")
letters_per_line = int(input("Enter the number of letters per line: "))

try:
    with open(file_name, 'w', encoding='utf-8') as file:
        alphabet = string.ascii_uppercase  # 'A' to 'Z'

        # Write letters in chunks of `letters_per_line`
        for i in range(0, len(alphabet), letters_per_line):
            line = ''.join(alphabet[i:i + letters_per_line])
            file.write(line + '\n')  # add newline at the end

    print(f"File '{file_name}' created successfully with {letters_per_line} letters per line.")

except IOError:
    print(f"Error: Could not write to file '{file_name}'.")

except ValueError:
    print("Error: Please enter a valid number for letters per line.")
```

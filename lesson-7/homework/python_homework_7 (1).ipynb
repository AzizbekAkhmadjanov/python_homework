âœ… 1. What is map()?

map(function, iterable) â€” applies a function to every element of a list (or any iterable) and returns a new map object (which can be converted to list).

âœ” Example 1 â€” Square each number
numbers = [1, 2, 3, 4, 5]

squares = list(map(lambda x: x**2, numbers))
print(squares)  # [1, 4, 9, 16, 25]

âœ” Example 2 â€” Convert strings to integers
strings = ["1", "2", "3"]

nums = list(map(lambda s: int(s), strings))
print(nums)  # [1, 2, 3]

âœ… 2. What is filter()?

filter(function, iterable) â€” keeps only those elements for which the function returns True.

âœ” Example 1 â€” Keep even numbers
numbers = [1, 2, 3, 4, 5, 6]

evens = list(filter(lambda x: x % 2 == 0, numbers))
print(evens)  # [2, 4, 6]

âœ” Example 2 â€” Keep strings longer than 3 characters
words = ["hi", "hello", "cat", "python"]

long_words = list(filter(lambda w: len(w) > 3, words))
print(long_words)  # ["hello", "python"]

âœ… Combined Example (map + filter)
âœ” Double the even numbers only
numbers = [1, 2, 3, 4, 5, 6]

result = list(
    map(lambda x: x * 2,
        filter(lambda x: x % 2 == 0, numbers)
    )
)

print(result)  # [4, 8, 12]

## 1. is_prime(n) funksiyasi
`is_prime(n)` funksiyasini hosil qiling (`n > 0`). Agar `n` soni tub bo'lsa `True`, aks holda `False` qiymat qaytarsin.

### Misollar:
- Kiritish:  
  `4`  
  Natija:  
  `False`  
  _(Izoh: `4` soni tub emas, chunki u `2` ga bo'linadi.)_

- Kiritish:  
  `7`  
  Natija:  
  `True`  
  _(Izoh: `7` soni faqat `1` va o'ziga bo'linadi, ya'ni tub son.)_


```python
def is_prime(n):
    if n < 2:
        return False

    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

print(is_prime(97))   # True
print(is_prime(4))    # False
```


## 2. digit_sum(k) funksiyasi
`digit_sum(k)` funksiyasini yozing, u `k` sonining raqamlari yig'indisini hisoblaydi.

### Misollar:
- **Kiritish:**  
  `24`  
  **Natija:**  
  `6`  
  _(Izoh: `24` sonining raqamlari yig'indisi: `2 + 4 = 6`.)_

- **Kiritish:**  
  `502`  
  **Natija:**  
  `7`  
  _(Izoh: `502` sonining raqamlari yig'indisi: `5 + 0 + 2 = 7`.)_


```python
def digit_sum(k):
    if k < 10:
        return k
    return k % 10 + digit_sum(k // 10)

print(digit_sum(24))   # 6
print(digit_sum(502))  # 7
```

## 3. Ikki sonining darajalari
Berilgan `N` sonidan oshmaydigan barcha 2 ning darajalarini (ya'ni, `2**k` shaklidagi sonlarni) chop etuvchi funksiyani yozing.

### Misol:
- **Kiritish:**  
  `10`  
  **Natija:**  
  `2 4 8`  
  _(Izoh: `10` dan kichik yoki teng bo'lgan `2` ning darajalari: `2, 4, 8`.)_

```python
def two_powers(n):
    result = []
    power = 2
    while power <= n:
        result.append(power)
        power *= 2
    return result

print(*two_powers(10))   # [2, 4, 8]
```
